# Helper Function

```python
def display_img(img):
    plt.imshow(img,cmap='gray')
    plt.show()
    
def plot_2img (img1, img2) :
    plt.figure(figsize = (20,5))
    
    plt.subplot(1,2,1)
    plt.imshow(img1, 'gray')
    
    plt.axis('off')
    
    plt.subplot(1,2,2)
    plt.imshow(img2, 'gray')
    
    plt.axis('off')
    
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR,category)#path to data
        class_num = CATEGORIES.index(category) #convert the string classification to number eg: 0 for bean,14 for tomato
        for img in os.listdir(path):# iterate though all image in train file
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                rgb_img=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
                
                resize_img=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
                rgb_resize=cv2.resize(rgb_img,(IMG_SIZE, IMG_SIZE))
                
                training_data.append([resize_img,class_num])
                training_dataRGB.append([rgb_resize,class_num])
            except Exception as e:
                pass # skip the image if there is error loading
            
            
def create_validation_data():
    for category in CATEGORIES:
        path=os.path.join(VALIDATION_DIR,category)#path to data
        class_num = CATEGORIES.index(category) #convert the string classification to number eg: 0 for bean,14 for tomato
        for img in os.listdir(path):# iterate though all image in train file
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                rgb_img=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
                
                resize_img=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
                rgb_resize=cv2.resize(rgb_img,(IMG_SIZE, IMG_SIZE))
                
                validation_data.append([resize_img,class_num])
                validation_dataRGB.append([rgb_resize,class_num])
            except Exception as e:
                pass # skip the image if there is error loading

def create_test_data():
    for category in CATEGORIES:
        path=os.path.join(TEST_DIR,category)#path to data
        class_num = CATEGORIES.index(category) #convert the string classification to number eg: 0 for bean,14 for tomato
        for img in os.listdir(path):# iterate though all image in train file
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                rgb_img=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
                
                resize_img=cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
                rgb_resize=cv2.resize(rgb_img,(IMG_SIZE, IMG_SIZE))
                
                test_data.append([resize_img,class_num])
                test_dataRGB.append([rgb_resize,class_num])
            except Exception as e:
                pass # skip the image if there is error loading
def reshape_data(X,y,rgb,rgb_y):
    X = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE, 1)
    y=np.array(y)
    rgb=np.array(rgb).reshape(-1,IMG_SIZE,IMG_SIZE, 3)
    rgb_y=np.array(rgb_y)
    return X,y,rgb,rgb_y

def append_Features(X,y,data):
    for features, label in data:
        X.append(features)
        y.append(label)
    return X,y
def save_data(name,data):
    pickle_out = open(f"Data/{name}","wb")
    pickle.dump(data,pickle_out)
    pickle_out.close()
```
