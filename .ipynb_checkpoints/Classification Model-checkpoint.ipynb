```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from sklearn.metrics import classification_report
import cv2
from sklearn.metrics import confusion_matrix 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dense,Dropout,Flatten,MaxPooling2D,Activation
```

# Helper Function

```python
CATEGORIES=['Bean','Bitter_Gourd','Bottle_Gourd','Brinjal','Broccoli','Cabbage','Capsicum','Carrot','Cauliflower','Cucumber','Papaya','Potato','Pumpkin','Radish','Tomato']
def display_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))
    title_fontsize = 16
    axis_fontsize = 12

    ax1.plot(range(1,epochs+1), history.history['loss'], label='Training loss')
    ax1.plot(range(1,epochs+1), history.history['val_loss'], label='Validation Loss')
    ax1.legend()
    ax1.set_xticks(range(1,epochs+1,3))
    ax1.set_title('Loss', fontsize=title_fontsize)
    ax1.set_xlabel('Epoch', fontsize=axis_fontsize)

    ax2.plot(range(1,epochs+1), history.history['accuracy'], label='Training Accuracy')
    ax2.plot(range(1,epochs+1), history.history['val_accuracy'], label='Validation Accuracy')
    ax2.legend()
    ax2.set_xticks(range(1,epochs+1,3))
    ax2.set_title('Accuracy', fontsize=title_fontsize)
    ax2.set_xlabel('Epoch', fontsize=axis_fontsize);

def display_confusion_matrix(y,predict):
    cm=confusion_matrix(y, predict.argmax(axis=1))

    ax= plt.subplot()
    sb.heatmap(cm, annot=True, fmt='g', ax=ax)
    ax.set_xticklabels(ax.get_xticklabels(), fontsize=8,rotation=90)
    ax.set_yticklabels(ax.get_yticklabels(), fontsize=8,rotation=0)
    plt.tight_layout()

    # labels, title and ticks
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels'); 
    ax.set_title('Confusion Matrix') 
    ax.xaxis.set_ticklabels(CATEGORIES)
    ax.yaxis.set_ticklabels(CATEGORIES)
    
def display_classification_report(y,predict):
    predictArr=[]
    for i in range(0,len(predict)):
        predictArr.append(np.argmax(predict[i]))  
    
    print(classification_report(y,predictArr , target_names=CATEGORIES))

def display_img(img):
    if(len(img[0][0])==1):
        plt.imshow(img,cmap='gray')
        plt.show()
    else:
        r,g,b = cv2.split(img)
        img_bgr = cv2.merge([b,g,r])
        plt.imshow(img_bgr)
        plt.show()
```

# Load data and normalise value

```python
X = pickle.load(open("Data/X.pickle","rb")) #images/feautures
y = pickle.load(open("Data/y.pickle","rb")) #classification/labels
X_rgb=pickle.load(open("Data/X_rgb.pickle","rb"))
y_rgb=pickle.load(open("Data/y_rgb.pickle","rb"))

validX = pickle.load(open("Data/validX.pickle","rb")) #images/feautures
validy = pickle.load(open("Data/validy.pickle","rb")) #classification/labels
validX_rgb=pickle.load(open("Data/validX_rgb.pickle","rb"))
validy_rgb=pickle.load(open("Data/validy_rgb.pickle","rb"))

testX = pickle.load(open("Data/testX.pickle","rb")) #images/feautures
testy = pickle.load(open("Data/testy.pickle","rb")) #classification/labels
testX_rgb=pickle.load(open("Data/testX_rgb.pickle","rb"))
testy_rgb=pickle.load(open("Data/testy_rgb.pickle","rb"))

#for i in range( 0,len(X) ):
X=np.divide(X,255.0)
X_rgb=np.divide(X_rgb,255.0)  # normalise value from 0-255 to 0-1

validX = np.divide(validX,255.0) 
validX_rgb=np.divide(validX_rgb,255.0) 

testX = np.divide(testX,255.0)
testX_rgb=np.divide(testX_rgb,255.0)
```

# 1st Model
## 2Conv2D layer for grayscale image

```python
#Create model for grayscale

model = Sequential()

model.add(Conv2D(64, (3,3), input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())


model.add(Dense(15))
model.add(Activation('softmax'))

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
model.summary()
```

```python
epochs=10
history = model.fit(X,y,batch_size=32,epochs=epochs, validation_data=(validX,validy),shuffle=True )# train the model
```

# Accuracy and Loss for grayscale image(10Epochs)
- Difference of Accuracy and Loss between Training and Validation differs highly around the end
- Signs of overfeeding occur around the 8th epoch of training

```python
display_training_history(history)
```

# Model 1 with lesser training cycles(5 epochs)
- To reduce overfitting
- From the graph below, the delta between training and validation is low

```python
#Create model for grayscale

model = Sequential()

model.add(Conv2D(64, (3,3), input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())


model.add(Dense(15))
model.add(Activation('softmax'))

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
epochs=5
history = model.fit(X,y,batch_size=32,epochs=epochs, validation_data=(validX,validy),shuffle=True )# train the model
display_training_history(history)
```

# 1st model against test images
- 80% accuracy

```python
model.evaluate(testX,testy)
grayscale_predict=model.predict([testX])
```

# Model results for grayscale image
- The model seem to be predicting poorly for cauliflower and pumpkin as it has high wrong prediction at 30
- In both category, the model predict the wrong classification most on Cabbage
- The gray scale images of Cabbage, Cauliflower and Pumpkin is round and since it has no colour data the model could easily mistaken one for the other

```python
display_confusion_matrix(testy,grayscale_predict)
```

```python
display_classification_report(testy,grayscale_predict)
```

```python
display_img(testX[3])#cabbage
display_img(testX[40])#cauliflower
display_img(testX[28])#pumpkin
```

# 2nd Model 
## 2Conv2d layer with RGB
- Including colour data into images may help model identify more features in the images

```python
model2 = Sequential()

model2.add(Conv2D(64, (3,3), input_shape=X_rgb.shape[1:]))
model2.add(Activation("relu"))
model2.add(MaxPooling2D(pool_size=(2,2)))

model2.add(Conv2D(64, (3,3)))
model2.add(Activation("relu"))
model2.add(MaxPooling2D(pool_size=(2,2)))

model2.add(Flatten())

model2.add(Dense(15))
model2.add(Activation('softmax'))

model2.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
epochs=10
history2 = model2.fit(X_rgb,y_rgb,batch_size=32,epochs=epochs,validation_data=(validX_rgb,validy_rgb),shuffle=True )# train the model
```

# Accuracy and Loss for RGB image model
- For this model, the Validation loss and accuracy seem to stabilise at 0.3 and is unable to be further reduce even after more training
- However it still manage to acheive a higher accuracy than the grayscale model with exact same model parameter at around 90%

```python
display_training_history(history2)
```

```python
model2.evaluate(testX_rgb,testy_rgb)
```

# Improving on RGB model using extra layer with dropout
- Adding 1 more 2DConv layer and 2 Dense layer with dropout to further reduce validation loss 

```python
model3 = Sequential()

model3.add(Conv2D(32, (3,3), input_shape=X_rgb.shape[1:]))
model3.add(Activation("relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))

model3.add(Conv2D(64, (3,3)))
model3.add(Activation("relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))

model3.add(Conv2D(128, (3,3)))
model3.add(Activation("relu"))
model3.add(MaxPooling2D(pool_size=(2,2)))

model3.add(Flatten())
model3.add(Dense(128))
model3.add(Dropout(rate=0.5))


model3.add(Dense(64))
model3.add(Dropout(rate=0.5))

model3.add(Dense(15))
model3.add(Activation('softmax'))

model3.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
epochs=10
history3 = model3.fit(X_rgb,y_rgb,batch_size=32,epochs=epochs, validation_data=(validX_rgb,validy_rgb),shuffle=True)# train the model
```

# Accuracy and Loss for modified RGB model
- With extra layer for the model, there is minimal delta between Training and Validation Accuracy/Loss
- 95% accuracy when model was tested against the test images

```python
display_training_history(history3)
```

```python
model3.evaluate(testX_rgb,testy_rgb)
predict_rgb=model3.predict([testX_rgb])
```

# Model results for RGB model
- The model did very well for carrots and capsicum probably due to their bright colour
- The addition of colour data did help the model improve its accuracy and reduce wrong classification

```python
display_confusion_matrix(testy_rgb,predict_rgb)
```

```python
display_classification_report(testy_rgb,predict_rgb)
```
